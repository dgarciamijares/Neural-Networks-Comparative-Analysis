As noted above, cell 5 in the CodeP2.4F23.ipynb code indicates how to predict values of α,
and ηsys for a submitted set of the input variables T1, γ, Qs. Below is a table of test data
(distinct from the training data). Note that you have to normalize the input data using the same median values as the trained model, insert the normalized values into an array, and submit the array to the model.predict function. The output results must be multiplied by the median values using in training to get physical output values.

```python
# Import useful packages
import json
import numpy as np

# ======================================
# Define test data arrays for x & y data
# ======================================
# T1(K), gamma, , qsol(kW):
xdata = [[318.0, 0.00, 500.0],
         [318.0, 0.00, 1500.0],
         [318.0, 0.00, 2500.0],
         [318.0, 0.25, 1500.0],
         [318.0, 0.50, 500.0],
         [318.0, 0.50, 1500.0],
         [318.0, 0.50, 2500.0],
         [303.0, 0.00, 1000.0],
         [303.0, 0.00, 2000.0],
         [303.0, 0.25, 1000.0],
         [303.0, 0.25, 2000.0],
         [303.0, 0.50, 1000.0],
         [303.0, 0.50, 2000.0],
         [288.0, 0.00, 500.0],
         [288.0, 0.00, 2500.0],
         [288.0, 0.25, 2500.0],
         [288.0, 0.50, 1500.0],
         [268.0, 0.00, 1500.0],
         [268.0, 0.25, 2000.0],
         [268.0, 0.50, 2500.0]]

# Alpha, EffSys:
ydata = [[35.13, 0.3808],
         [47.46, 0.3930],
         [73.12, 0.4061],
         [66.34, 0.4098],
         [63.09, 0.4154],
         [85.23, 0.4197],
         [131.32, 0.4242],
         [38.99, 0.4012],
         [53.80, 0.4136],
         [54.51, 0.4215],
         [75.22, 0.4290],
         [70.04, 0.4337],
         [96.65, 0.4382],
         [33.45, 0.4091],
         [60.80, 0.4334],
         [85.044, 0.4477],
         [77.56, 0.4516],
         [40.68, 0.4383],
         [65.24, 0.4628],
         [98.23, 0.4760]]

# ======================================================
# Load median values from json file of the training data
# ======================================================
with open("x_medians.json", "r") as x_medians_file:
    x_medians_dict = json.load(x_medians_file)    
with open("y_medians.json", "r") as y_medians_file:
    y_medians_dict = json.load(y_medians_file)
# Access the data from the dictionaries
x_medians = x_medians_dict["data"]
y_medians = y_medians_dict["data"]
# Transforming the lists to arrays
x_medians = np.array(x_medians)
y_medians = np.array(y_medians)

# ========================================
# Normalization of x & y test data
# ========================================
n_xdata = xdata/x_medians
n_ydata = ydata/y_medians

# ===============================================
# Use of the load/unload capability of json files
# ===============================================
# Convert n_xdata and n_ydata to dictionaries
n_xdata_dict = {"data": n_xdata.tolist()}
n_ydata_dict = {"data": n_ydata.tolist()}

# Convert the dictionaries to JSON format
n_xdata_json = json.dumps(n_xdata_dict, indent=4)
n_ydata_json = json.dumps(n_ydata_dict, indent=4)

# Save the JSON data to files if needed
with open("n_xdata_test.json", "w") as xdata_file:
    xdata_file.write(n_xdata_json)
with open("n_ydata_test.json", "w") as ydata_file:
    ydata_file.write(n_ydata_json)

# ===============================================
# Check of data dimensions and normalized values
# ===============================================
print(x_medians) # Tmed, gamed, qsmed
print(y_medians) # almed, efmed
# Dimension of xdata & ydata
print("xdata rows ",len(xdata)," and columns ", len(xdata[0]))
print("ydata rows ",len(ydata)," and columns ", len(ydata[0]))
# Print original data
print("xdata")
print(xdata)
print("ydata")
print(ydata)
# Dimension of normalized xdata & ydata
print("n_xdata rows ",len(n_xdata)," and columns ", len(n_xdata[0]))
print("n_ydata rows ",len(n_ydata)," and columns ", len(n_ydata[0]))
# Print normalized data
print("n_xdata")
print(n_xdata)
print("n_ydata")
print(n_ydata)
```
