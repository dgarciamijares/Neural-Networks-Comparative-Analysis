these starting values with those for the original starting values (that were the same as the first <br>

```python
import matplotlib.pyplot as plt
import numpy as np

# Data from 1st principles and Keras best results
y_data = [30.97, 32.30, 31.50, 30.91, 32.50, 31.40, 35.59, 46.40]
first_principles = [33.00, 33.59, 33.89, 33.28, 33.59, 33.89, 38.64, 56.17]
keras = [31.72, 32.31, 32.60, 31.91, 32.31, 32.79, 35.54, 47.62]

# Calculate logarithms (base 10) of the data
log_y_data = np.log10(y_data)
log_first_principles = np.log10(first_principles)
log_keras = np.log10(keras)

# Create log-log plots
plt.figure(figsize=(8, 6))

# Plot y_data as a function of 1st_principles
plt.loglog(log_first_principles, log_y_data, marker='o', linestyle='', color='b', label='1st Principles')
# Plot y_data as a function of Keras
plt.loglog(log_keras, log_y_data, marker='o', linestyle='', color='g', label='Keras')

# Fitting
# Perfect fit
x_range_min = min(min(log_first_principles), min(log_keras))
x_range_max = max(max(log_first_principles), max(log_keras))
x_range = np.linspace(x_range_min, x_range_max, 100)
plt.loglog(x_range, x_range, color='r', linestyle='--', label='Perfect Fit')

# 1st principles predicted values
coefficients_fp = np.polyfit(log_first_principles, log_y_data, 1)
fit_line_fp = np.poly1d(coefficients_fp)
plt.loglog(log_first_principles, fit_line_fp(log_first_principles), color='b', linestyle='--', label='Fit - 1st Principles')

# Keras predicted values
coefficients_k = np.polyfit(log_keras, log_y_data, 1)
fit_line_k = np.poly1d(coefficients_k)
plt.loglog(log_keras, fit_line_k(log_keras), color='g', linestyle='--', label='Fit - Keras')

# Labeling and styling
plt.xlabel('y predicted (Log Scale)')
plt.ylabel('y data (Log Scale)')
plt.title('Log-Log Plot of Predicted vs. Data')
plt.grid(False)
plt.legend()

# Show the plot
plt.show()

print("Fit Equation (1st Principles): y = {:.2f}x + {:.2f}".format(coefficients_fp[0], coefficients_fp[1]))
print("Fit Equation (Keras): y = {:.2f}x + {:.2f}".format(coefficients_k[0], coefficients_k[1]))
```
